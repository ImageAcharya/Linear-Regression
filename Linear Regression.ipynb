```python
#Best-Fit Line Using Gradient Descent
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data 
X = np.array([1, 2, 3, 4, 5])   #Example- Hours of study
Y = np.array([2, 4.1, 6.2, 8.3, 10.5]) #Scores achieved after the respective study hours
```

```python
# Parameters
learning_rate = 0.0001 
n_iterations = 1000
m = len(X)  

theta_0 = 0
theta_1 = 0
theta_2 = 0  


for i in range(n_iterations):
    Y_pred = theta_0 + theta_1 * X + theta_2 * X**2
    
    d_theta_0 = -2 * np.sum(Y - Y_pred)
    d_theta_1 = -2 * np.sum((Y - Y_pred) * X)
    d_theta_2 = -2 * np.sum((Y - Y_pred) * X**2)
    
    # Update
    theta_0 = theta_0 - learning_rate * d_theta_0
    theta_1 = theta_1 - learning_rate * d_theta_1
    theta_2 = theta_2 - learning_rate * d_theta_2
```

```python
# Plotting the data points
X_range = np.linspace(min(X), max(X), 100)  
Y_pred = theta_0 + theta_1 * X_range + theta_2 * X_range**2

plt.scatter(X, Y, color="yellow", label="Data points")
plt.plot(X_range, Y_pred, color="green", label="Best fit curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Print the coefficients
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
print(f"theta_2: {theta_2}")
```
